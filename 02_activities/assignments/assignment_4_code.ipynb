## Police Annual Statistical Report - Victims of Crimes
https://open.toronto.ca/dataset/police-annual-statistical-report-victims-of-crime/


The dataset includes information on identified victims of crimes against the person in Toronto.
 
Key columns include REPORT_YEAR, CATEGORY, SUBTYPE, ASSAULT_SUBTYPE, SEX, AGE_GROUP, AGE_COHORT, and COUNT_.

```python
import pandas as pd

# Load the dataset
file_path = 'Victims of Crime.csv'
df = pd.read_csv(file_path)

# Display basic information about the dataset to understand its structure
df.info(), df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization 1: Crime Distribution by Year and Category
plt.figure(figsize=(12, 6))

crime_trend = df.groupby(['REPORT_YEAR', 'CATEGORY'])['COUNT_'].sum().reset_index()

sns.lineplot(data=crime_trend, x='REPORT_YEAR', y='COUNT_', hue='CATEGORY', marker='o')

plt.title('Crime Distribution by Year and Category')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Category')
plt.grid(True)
```

```python

# Visualization 2: Age Cohort and Sex Distribution of Victims
plt.figure(figsize=(12, 6))

age_sex_distribution = df.groupby(['AGE_COHORT', 'SEX'])['COUNT_'].sum().reset_index()

sns.barplot(data=age_sex_distribution, x='AGE_COHORT', y='COUNT_', hue='SEX')

plt.title('Age Cohort and Sex Distribution of Crime Victims')
plt.xlabel('Age Cohort')
plt.ylabel('Number of Victims')
plt.legend(title='Sex')
plt.grid(True)
```
