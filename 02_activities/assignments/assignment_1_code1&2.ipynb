```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#set random seed
np.random.seed(613)

# create array where value from 0 to 49
x = np.arange(50) 

# create a random array contains 50 number where the value is from 0 to 100
y = np.random.randint(0, 100, 50) 
```

```python
# get fig and axes from subplots function

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
# explore differnt ax style

fig, ax = plt.subplots(figsize=(5, 3))
# ax.bar(x,y)
# ax.barh(x,y)
# ax.plot(x,y)
ax.hist(y)
```

```python
# set lables in ax

fig, ax = plt.subplots(figsize=(5, 3))
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
ax.scatter(x,y)
```

```python
# set font properties for lables in the axes 

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# Modifying data points

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,
y,
marker='*', # marker argument
color = "indigo") # color argument

# https://matplotlib.org/stable/api/markers_api.html
```

```python
# adjust the style and width of the line connecting our points

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
y,
marker='*',
color = 'indigo',
linestyle = '--', # linestyle argument
linewidth = 2) # linewidth argument
```

```python
# modify colour of data points

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
markersize = 12, # markersize argument
color = '#7425b9',
linestyle = '--',
linewidth = 2,
markeredgecolor = '#fa9359', # markeredgecolor argument
markerfacecolor = '#000000' ) # markerfacecolor argument

ax.grid(axis = 'y')
ax.grid(color = "blue", linewidth = 2, linestyle = '-.')
```

```python
# Exploring more examples

np.random.seed(42)

x = np.arange(10) 

y = np.random.randn(3, 10) 

fig, ax = plt.subplots(figsize=(5, 3))

ax.stackplot(x, y, labels=['A','B','C'])
ax.legend(loc ='upper right')
ax.set_title('Exploring stackplot') 
```
